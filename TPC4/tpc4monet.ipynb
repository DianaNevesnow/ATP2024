```python
import random

# Mostrar o menu
def mostrar_menu():
    print("\n### MENU ###")
    print("(1) Criar Lista")
    print("(2) Ler Lista")
    print("(3) Soma")
    print("(4) Média")
    print("(5) Maior")
    print("(6) Menor")
    print("(7) Está Ordenada por ordem crescente?")
    print("(8) Está Ordenada por ordem decrescente?")
    print("(9) Procurar um elemento")
    print("(0) Sair")
    print("Escolha uma opção:", end=" ")

# Criar uma lista aleatória
def criar_lista():
    size = int(input("Quantos números quer na lista? "))
    list = [random.randint(1, 100) for c in range(size)]
    print("Lista criada:", list)
    return list

# Ler uma lista do utilizador
def ler_lista():
    list=[]
    print("Introduza os números (escreva 'fim' para terminar):")
    value = input("Número: ")
    while value != "fim":
        list.append(int(value))
        value = input("Número: ")
    print("Lista criada:", list)
    return list

# Calcular a soma dos elementos
def calcular_soma(list):
    if not list:
        print("A lista está vazia.")
    else:
        soma = sum(list)
        print("Soma dos elementos:", soma)

# Calcular a média dos elementos
def calcular_media(list):
    if not list:
        print("A lista está vazia. Não é possível calcular a média.")
    else:
        media = sum(list) / len(list)
        print("Média dos elementos:", media)

# Encontrar o maior elemento
def encontrar_maior(list):
    if not list:
        print("A lista está vazia.")
    else:
        maior = max(list)
        print("Maior elemento:", maior)

# Encontrar o menor elemento
def encontrar_menor(list):
    if not list:
        print("A lista está vazia.")
    else:
        menor = min(list)
        print("Menor elemento:", menor)

# Verificar se a lista está ordenada em ordem crescente
def esta_ordenada_crescente(list):
    if len(list) <= 1 or all(list[i] <= list[i + 1] for i in range(len(list) - 1)):
        print("Sim")
    else:
        print("Não")

# Verificar se a lista está ordenada em ordem decrescente
def esta_ordenada_decrescente(list):
    if len(list) <= 1 or all(list[i] >= list[i + 1] for i in range(len(list) - 1)):
        print("Sim")
    else:
        print("Não")

# Procurar um elemento na lista
def procurar_elemento(list):
    element = int(input("Qual elemento quer procurar? "))
    position = -1
    for i, value in enumerate(list):
        if value == element:
            position = i
    if position == -1:
        print("Elemento não encontrado.")
    else:
        print(f"Elemento encontrado na posição {position}.")

# Função principal
def main():
    list = []
    while True:
        mostrar_menu()
        option = input()

        if option == "1":
            list = criar_lista()
        elif option == "2":
            list = ler_lista()
        elif option == "3":
            calcular_soma(list)
        elif option == "4":
            calcular_media(list)
        elif option == "5":
            encontrar_maior(list)
        elif option == "6":
            encontrar_menor(list)
        elif option == "7":
            esta_ordenada_crescente(list)
        elif option == "8":
            esta_ordenada_decrescente(list)
        elif option == "9":
            procurar_elemento(list)
        elif option == "0":
            print("Lista final:", list)
            print("Até breve!")
            return
        else:
            print("Opção inválida. Tente novamente.")

# Chamar a função principal
main()
```
