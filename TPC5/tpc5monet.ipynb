```python
import time
cinema=[]

def menu():
    print("----------------------\n(1)Criar sala \n (2)Remover sala \n (3)Listar salas \n (4)Lugar disponivel \n (5)Vender bilhete \n (6)Consultar salas \n (0)Sair/Reset")
    option=int(input("Escolha uma função:"))
    if option==1:
        inserirSala(cinema)
    elif option==2:
        removerSala(cinema)
    elif option==3:
        listarSalas(cinema)
    elif option==4:
        film=input("Introduza o nome do filme:")
        lugar= int(input("Introduza o lugar:"))
        print("Lugar disponivel?",lugarDisponivel(cinema,film,lugar))
    elif option==5:
        film= input("Introduza o nome do filme:")
        lugar= int(input("Introduza o lugar:"))
        vendebilhete(cinema,film,lugar)
    elif option==6:
        listardisponibilidadesala(cinema)
    elif option==0:
        print ("reset efetuado com sucesso")
        print("Salas:", cinema)
menu()


def inserirSala(cinema):
    film = input("Introduza o nome do filme:")
    for f in cinema:
        nlugares, sold, nome = f
        if nome == film: 
            print("Filme já existe.")
            menu()
    nlugares = int(input("Introduza o numero de lugares da sala:"))
    sold =[]
    cinema.append((nlugares, sold, film))
    print("A sala foi criada com sucesso!")
    print("Salas:", cinema)
    menu()


def removerSala(cinema):
    index=int(input("Introduza o indice da sala:"))
    cinema.pop(index)
    print("Sala removida com sucesso")
    print("salas:", cinema)
    menu()

def listarSalas(cinema):
    print("Sala     Filme")
    print("--------------")
    for p in cinema:
        nlugares, sold, film = p
        print("Sala#{0}".format(cinema.index(p)),"",film)
    time.sleep(1)
    menu()
        
def lugarDisponivel (cinema, film, lugar):
    for p in cinema:
        nlugares, sold, nome = p
        if film == nome:
            if lugar in sold:
                res=False
            else: 
                res=True
        else:
            res=False
    return res                

def vendebilhete (cinema, film, lugar):
    if lugarDisponivel(cinema, film, lugar):
        for p in cinema:
            nlugares, sold, nome =p
            if film==nome:
                sold.append(lugar)
        print("Lugar vendido com sucesso")
        print("salas:",cinema)
    else:
        print("lugar disponivel!")
    menu()

def listardisponibilidadesala(cinema):
    print ("Sala    Filme   Lugares disponiveis")
    print("------------------------------------")
    for p in cinema:
        nlugares, sold, film = p
        print("Sala#{0}".format(cinema.index(p)),"",film,"",nlugares - len(sold))
    time.sleep(1)
    menu()
```
